```python
import sys

if "../" not in sys.path:
  sys.path.append("../")

from lib.envs.cliff_walking import CliffWalkingEnv
from lib import plotting
from agents import QLearningAgent, EVSarsaAgent
import numpy as np
```

```python
"""
o  o  o  o  o  o  o  o  o  o  o  o
o  o  o  o  o  o  o  o  o  o  o  o
o  o  o  o  o  o  o  o  o  o  o  o
x  C  C  C  C  C  C  C  C  C  C  T
"""

env_shape = (4, 12)
start_position = (3, 0)
end_positions = [(3, 11)]
cliff = tuple((3, i+1) for i in range(10))

env = CliffWalkingEnv(env_shape, start_position, end_positions, cliff)
print("Environment")
env.render()

n_actions = env.action_space.n
agent_ql = QLearningAgent(alpha=0.5, epsilon=0.1, discount=0.99, n_actions=n_actions)
agent_sarsa = EVSarsaAgent(alpha=0.5, epsilon=0.1, discount=0.99, n_actions=n_actions)
agent_ql.train(env, n_episodes=3000, t_max=10**3, verbose=True, verbose_per_episode=500)
agent_sarsa.train(env, n_episodes=3000, t_max=10**3, verbose=True, verbose_per_episode=500)

print("\nQ Learning Policy")
plotting.draw_policy(env, agent_ql)
print("\nSARSA Policy")
plotting.draw_policy(env, agent_sarsa)
plotting.plot_stats(agent_ql, agent_sarsa)
```
